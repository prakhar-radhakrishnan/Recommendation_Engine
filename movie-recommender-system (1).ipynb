```python
import numpy as np 
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
#checking if there are some missing datas.....three movies dont have overview

movies.isnull().sum()
```

```python
#since only 3, and hence not a bog no., so we drop/ignore 
movies.dropna(inplace=True)
```

```python
#need only keywords, do by helper function
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L        
```

```python
#since the columns contains in form of strings, hence convert fncn not run directly....
#hence need to convert these string of list into list 
#can do by module ...
import ast 
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#SAME FOR keywords,...what we did with genres ,same
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#crew..only need where job value is director
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L    
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#the colm of overview is a string...so convert in list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#concatinating all the lists
#but before that , need to remove spaces b/w the words...bcz spaces can act as seperate entities..prblm in name...eg....2 "SAM"
#so we will make "sameworthinton" a single entity
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
#making a new column of TAG
movies['tags'] = movies['overview']+movies['genres']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
#removing the other columns 
new_df = movies[['movie_id','title','tags']]
```

```python
#COnvertinf the lists of tags in strings ..i.e give spaces
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
#calculating distance ....by cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]


#gives the distance of first movies woith all other mvoies
```

```python
#making a fucntion so if we give one movie , it returns 5 similar movies
#new_df.iloc[i[0]].title
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
  
```

```python
recommend('Iron Man')
```
